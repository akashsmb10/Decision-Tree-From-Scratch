```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
print(data.DESCR)
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df
```

```python
data.target
```

```python
x=df
y=data.target
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion="entropy")
classifier
```

```python
classifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,30))
tree.plot_tree(classifier,filled=True)
```

```python
plt.show()
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# Postpruning >> after making complete decision tree you have to cut it to avoid overfitting

# Observation 
# observe the decisioon tree plot, you will realise the last split is actually ovrerfitting
# Dividing the data in such a way that each node is 100 percent pure leading to low number of data points
# so reduce the depth of decision tree
```

```python
classifier=DecisionTreeClassifier(criterion="entropy",max_depth=2)
classifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,30))
tree.plot_tree(classifier,filled=True)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# Accuracy is sanme , Earlier uit was an Overftting model
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# preprunning using hyperparameter tunning
parameter={
    'criterion':["gini","entropy","log_loss"],
    'splitter' :["best","random"],
    'max_depth':[1,2,3,4,5,6,7,10],
    'max_features':["sqrt","log2","auto"]
}
```

```python
from sklearn.model_selection import GridSearchCV
clf=DecisionTreeClassifier()
model=GridSearchCV(clf,param_grid=parameter,cv=5,scoring="accuracy",verbose=3)
model
```

```python
model.fit(x_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred=model.best_estimator_.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```
